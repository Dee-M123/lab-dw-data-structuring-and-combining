```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_data.csv") #from previous labs
df.head()
```

```python
df.shape
```

```python
df_1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_1.head(2)
```

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
df_2.head()
```

```python
df_1.columns = df_1.columns.str.lower().str.replace(" ", "_")
df_1 = df_1.rename(columns={'st': 'state'})
df_1['gender'] = df_1['gender'].str.strip().str.lower()
df_1['gender'] = df_1['gender'].replace({'m': 'M','male': 'M','f': 'F','female': 'F','femal': 'F'})
```

```python
state_mapping = {"AZ": "Arizona","Cali": "California","WA": "Washington"}

df_1['state'] = df_1['state'].replace(state_mapping)
df_1['education'] = df_1['education'].replace({'Bachelors': 'Bachelor'})

df_1['customer_lifetime_value'] = (df_1['customer_lifetime_value'].astype(str).str.replace('%', '', regex=False).astype(float))
```

```python
df_1["number_of_open_complaints"] = (
    df["number_of_open_complaints"]
    .astype(str)              # ensure string for split
    .str.split("/")           # split into parts
    .str[1]                   # take the middle value
)

df_1["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"],errors="coerce")

df_1= df_1.drop_duplicates()

df_1 = df_1.copy()
df_1["customer"]=df_1["customer"].dropna()

df_1 = df_1.copy()
df_1["education"]=df_1["education"].fillna("unkown")

df_1 = df_1.copy()
df_1["gender"] = df_1["gender"].fillna("unkown")

df_1.copy()

df_1[["state","policy_type","vehicle_class"]]=df_1[["state","policy_type","vehicle_class"]].fillna("missing")

df_1.copy
df_1 = df_1.dropna()
```

```python
df_1.copy()
df_1[["customer_lifetime_value","income","monthly_premium_auto","number_of_open_complaints","total_claim_amount"]]=df_1[["customer_lifetime_value","income","monthly_premium_auto","number_of_open_complaints","total_claim_amount"]].fillna(df[["customer_lifetime_value","income","monthly_premium_auto","number_of_open_complaints","total_claim_amount"]].mean())
df_1.head()
```

```python
df_2.columns = df_2.columns.str.lower().str.replace(" ", "_")
df_2 = df_2.rename(columns={'st': 'state'})
df_2['gender'] = df_2['gender'].str.strip().str.lower()
df_2['gender'] = df_2['gender'].replace({'m': 'M','male': 'M','f': 'F','female': 'F','femal': 'F'})
```

```python
state_mapping = {"AZ": "Arizona","Cali": "California","WA": "Washington"}

df_2['state'] = df_2['state'].replace(state_mapping)
df_2['education'] = df_2['education'].replace({'Bachelors': 'Bachelor'})

df_2['customer_lifetime_value'] = (df_2['customer_lifetime_value'].astype(str).str.replace('%', '', regex=False).astype(float))
```

```python
df_2["number_of_open_complaints"] = (
    df["number_of_open_complaints"]
    .astype(str)              # ensure string for split
    .str.split("/")           # split into parts
    .str[1]                   # take the middle value
)

df_2["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"],errors="coerce")

df_2= df_2.drop_duplicates()

df_2 = df_2.copy()
df_2["customer"]=df_2["customer"].dropna()

df_2 = df_2.copy()
df_2["education"]=df_2["education"].fillna("unkown")

df_2 = df_2.copy()
df_2["gender"] = df_2["gender"].fillna("unkown")

df_2.copy()

df_2[["state","policy_type","vehicle_class"]]=df_2[["state","policy_type","vehicle_class"]].fillna("missing")

df_2.copy
df_2 = df_2.dropna()
```

```python
df_2.head(2)
```

```python
df_all = pd.concat([df, df_1, df_2], axis=0, ignore_index=True)
df_all.head()
```

```python
df_all = df_all.drop_duplicates()
df_all.shape
```
